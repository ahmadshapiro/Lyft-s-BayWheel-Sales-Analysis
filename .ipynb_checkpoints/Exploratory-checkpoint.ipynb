### Importing libraries 

```python
import pandas as pd
import seaborn as sns 
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib.patches as mpatches
import folium
%matplotlib inline
```

```python
df=pd.read_csv("2017-fordgobike-tripdata.csv")
df.head()
```

# Assessing Data and Cleaning Data 

```python
df.head()
```

```python
df.info()
```

The `start_time` and `end_time` need to be casted to date and time instead of string to ease the use of them in the analysis

```python
df['end_time']=pd.to_datetime(df['end_time'])
df['start_time']=pd.to_datetime(df['start_time'])
df.info()
```

**As we can see above there's no missing data in the columns , so we can start the analysis**

# **Exploratory Data Analysis**

**Since the bike IDs are unique to each bike, let's see how many bike does BayWheels has**

```python
df.bike_id.nunique()
```

**Let's see the top 10 rented bikes**

```python
top_10_bikes=df.bike_id.value_counts()[0:10]
plt.figure(figsize=(10,5))
sns.barplot(x=top_10_bikes.index.values,y=top_10_bikes.values,color=sns.color_palette('muted')[0],order=top_10_bikes.index.values);
plt.xlabel("Bike ID");
plt.ylabel("Number of time rented");
plt.title("Top 10 Rented Bikes");
```

**Let's check the distribution of the renteal duration in minutes**

```python
duration_minutes=df.duration_sec/60
```

```python
plt.figure(figsize=(10,5))
bin_edges = 10 ** np.arange(np.log10(duration_minutes.min()), np.log10(duration_minutes.max())+0.1, 0.1)
plt.hist(duration_minutes, bins = bin_edges,color=sns.color_palette('muted')[0]);
plt.xscale('log');
tick_locs = [1,3,10, 30, 100, 300, 1000];
plt.xticks(tick_locs, tick_locs);
plt.xlabel("Duration (Min)");
plt.ylabel("Count");
plt.title("Rental Duration");
```

We can see from the above plot that the rental duration median is around 10 minutes and it's normaly distributed with many outliers

**Lets check Top 10 stations which have the most rentals**

```python
top_10_stations=df[df.start_station_name.isin(df.start_station_name.value_counts()[0:10].index)].start_station_name
plt.figure(figsize=(10,5))
sns.countplot(y=top_10_stations,color=sns.color_palette('muted')[0],order=top_10_stations.value_counts().index.values);
plt.ylabel("Station");
plt.xlabel("Count");
plt.title("Top 10 Rental Stations");
```

**Lets check Top 10 returning stations**

```python
top_10_stations_return=df[df.end_station_name.isin(df.end_station_name.value_counts()[0:10].index)].end_station_name
plt.figure(figsize=(10,5))
sns.countplot(y=top_10_stations_return,color=sns.color_palette('muted')[0],order=top_10_stations_return.value_counts().index.values);
plt.ylabel("Station");
plt.xlabel("Count");
plt.title("Top 10 Returning Stations");
```

**Let's see the renting hours**

```python
df['start_hour']=df.start_time.apply(lambda x:x.hour)
```

```python
labels=["Night",'Morning',"After Noon","Evening"]
bins=[0,6,12,17,23]
```

```python
df['start_hour_labeld']=pd.cut(df.start_hour,bins=bins,labels=labels)
```

```python
plt.figure(figsize=(10,5))
sns.countplot(df.start_hour,color=sns.color_palette('muted')[0],);
plt.ylabel("Count");
plt.xlabel("Hour");
plt.title("Renting Hours");
```

As we can see in the above plot that the rush hour for renting is 8:00 AM and 

**Let's see the different types of users**

```python
plt.figure(figsize=(10,5))
sns.countplot(df.user_type,color=sns.color_palette('muted')[0])
plt.ylabel("Count");
plt.xlabel("Type");
plt.title("Users Types");
```

```python
plt.figure(figsize=(20,10))
ax=sns.countplot(df.start_hour_labeld);
plt.ylabel("Count",fontsize=15);
plt.xlabel("Day time",fontsize=15);
plt.title("Rush Hours",fontsize=30);
```

## **Do the top 10 renting stations have different rush hours ?**

```python
top_10_stations_df=df[df.start_station_name.isin(df.start_station_name.value_counts()[0:10].index)]
```

```python
plt.figure(figsize=(30,20))
ax=sns.countplot(y=top_10_stations_df.start_station_name,
                 hue=top_10_stations_df.start_hour_labeld);
plt.ylabel("Rental Station",fontsize=25);
plt.xlabel("Count",fontsize=25);
plt.title("Rush Hours",fontsize=30);
plt.xticks(fontsize=15)
plt.yticks(fontsize=25)
plt.legend(fontsize=25);
```

As we can see above both `Powel St BART Station` and `The Embarcadero at Sansome St` have rush hours at the after noon rather than the morning

## **Which days has more renting ?**

```python
plt.figure(figsize=(20,10))
sns.countplot(df.start_time.dt.day_name(),
              order=df.start_time.dt.day_name().value_counts().index);
plt.xlabel("Day of Week",fontsize=15);
plt.ylabel("Count",fontsize=15);
plt.title("Renting on Days of Week",fontsize=25);
```

We can see the huge difference in Saturdays and Sundays, they're much lower than the rest of the week 

## **Do people rent more time on specific days ?**

```python
plt.figure(figsize=(20,10));
ax=sns.boxenplot(x=df.start_time.dt.day_name(),
                 y=duration_minutes);
plt.xticks(rotation=30);
plt.yscale('log');
plt.yticks(tick_locs,tick_locs);
plt.ylabel("Durations (Min)",fontsize=15);
plt.xlabel("Day of Week",fontsize=15);
plt.title("Durations of rental in each day",fontsize=25);
```

As we can see on Sundays and Saturday the median is more than the rest of the week , which means that people rents in more duration on those days 

**We can conclude from the previous two plots , that people tend to rent less on weekends , but when they do , they rent for a long time**

## **Does Start Station affects the renting duration?**

```python
top_10_stations_df=top_10_stations_df.copy()
top_10_stations_df['duration_min']=top_10_stations_df.duration_sec/60
my_order=top_10_stations_df.groupby(by=["start_station_name"])["duration_min"].median().sort_values(ascending=False).index
plt.figure(figsize=(20,10))
ax=sns.boxplot(y=top_10_stations_df.start_station_name,
                x=top_10_stations_df.duration_min,
                order=my_order,
                color=sns.color_palette('muted')[0]);
plt.xscale('log');
plt.xticks(tick_locs,tick_locs,fontsize=20);
plt.xlabel("Durations (Min)",fontsize=15);
plt.ylabel("Start Station",fontsize=15);
plt.yticks(fontsize=20)
plt.title("Durations of rental in TOP 10 Start Station",fontsize=25);
```

As we can see that "The Embarcadero at Sansome St" has a the biggest median and interquantile range , which means it has more rental duration than the other stations

## **Let's Check the weekly sales(Rentals)**

```python
rentals_by_date=df.groupby(df.start_time.dt.date)['duration_sec'].count().to_frame()
rentals_by_date_rolled=rentals_by_date.rolling(7,min_periods=1).mean()
rentals_by_date.rename(columns={'start_time':'date','duration_sec':'rentals'},inplace=True)
rentals_by_date_rolled.reset_index(inplace=True)
rentals_by_date_rolled.rename(columns={'start_time':'date','duration_sec':'rentals'},inplace=True)
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x='date',y='rentals',data=rentals_by_date_rolled,color='black');
plt.xlabel("Date",fontsize=15);
plt.ylabel("Number of Rentals",fontsize=15);
plt.title("Weekly Average for the number of rentals",fontsize=20);
```

We can see a huge decline at December, the reason is the holidays , Because people prefer staying with families rather than cycling

## **Let's see relationship between the user type and the number of rides and the week of day** 

```python
plt.figure(figsize=(20,10))
sns.countplot(y=df.start_time.dt.day_name(),hue=df.user_type);
plt.xlabel("Number of Rides",fontsize=15)
plt.ylabel("Day Of Week",fontsize=15)
plt.legend(fontsize=15)
plt.title("User type VS number of rides in week days",fontsize=25);
```

We can see from the above plots that the number of customers increase on Sundays and Saturdays

## Is there any special days the cutomers ride more than subscribers?

```python
duration_minutes=df.duration_sec/60
plt.figure(figsize=(20,10))
sns.barplot(x=duration_minutes,y=df.start_time.dt.day_name(),hue=df.user_type);
plt.xlabel("Mean Duration of Rides (MIN)",fontsize=15)
plt.ylabel("Day Of Week",fontsize=15)
plt.legend(fontsize=15)
plt.title("User type VS Duration for Each week of day",fontsize=25);
```

**We can conclude from the above two plots that subscribers make a huge number of rides but with a small duration compared to the Customers** 
</br> **We can say that the subscribers renting the bicycle for a certain reason that's why thier mean is approximatly the same, in contrast , the costumers are just having rides..** 

## **Let's check the rush hours at each day for both type of user**

```python
df['day_of_week']=df.start_time.dt.day_name()
df['duration_minutes']=df.duration_sec/60
grouped = df.groupby(['user_type','day_of_week','start_hour']).mean()['duration_minutes']
grouped = grouped.reset_index(name = 'duration_min_avg')
grouped
custs=grouped[grouped['user_type']=="Customer"][['day_of_week','start_hour','duration_min_avg']]
custs=custs.pivot(index = 'day_of_week', columns = 'start_hour', 
                            values = 'duration_min_avg')
subsc=grouped[grouped['user_type']=="Subscriber"][['day_of_week','start_hour','duration_min_avg']]
subsc=subsc.pivot(index = 'day_of_week', columns = 'start_hour', 
                            values = 'duration_min_avg')
subsc
```

```python
plt.figure(figsize=(20,10));
sns.heatmap(custs,annot=True,fmt='.1f',cmap="Blues",cbar_kws = {'label':'Average Rides Duration (min)'});
plt.xlabel("Hour",fontsize=15);
plt.ylabel("Day Of Week",fontsize=15);
plt.xticks(fontsize=10);
plt.yticks(fontsize=10);
plt.title("Customers Average Ride Durations",fontsize=20);
```

**As we can see from the above heat mapes that rush hours from customers are in the morning as we conlcuded before**

```python
plt.figure(figsize=(20,10));
sns.heatmap(subsc,annot=True,fmt='.1f',cmap="Blues",cbar_kws = {'label':'Average Rides Duration (min)'});
plt.xlabel("Hour",fontsize=15);
plt.ylabel("Day Of Week",fontsize=15);
plt.xticks(fontsize=10);
plt.yticks(fontsize=10);
plt.title("Subscribers Average Ride Durations",fontsize=20);
```
